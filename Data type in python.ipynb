# Python Data Types for Beginners

In Python, every piece of data is an object. Understanding Python’s built-in data types is fundamental for writing efficient, readable, and bug-free programs.

## 1. Numeric Types

Python has three main numeric data types:

1. Integer (int) – Whole numbers (e.g., 0, 1, -5, 999)
2. Float (float) – Numbers with decimal points (e.g., 3.14, 2.7e3)
3. Complex (complex) – Numbers with real and imaginary parts (e.g., 2 + 3j)

### 1.1 Integers

Integers are whole numbers without a fractional component.
Examples: -10, 0, 42

```python
# Example of integers
num1 = 42
num2 = -7
num3 = 0

print("num1:", num1, type(num1))
print("num2:", num2, type(num2))
print("num3:", num3, type(num3))
```

### 1.2 Float

Float numbers are numbers that contain decimal points 
or are written in exponential form.

Examples: 3.14, -2.5, 0.0, 2.7e3

```python
# Example of floats
num1 = 3.14
num2 = -2.5
num3 = 0.0
num4 = 2.7e3   # exponential form (2.7 × 10^3)

print("num1:", num1, type(num1))
print("num2:", num2, type(num2))
print("num3:", num3, type(num3))
print("num4:", num4, type(num4))
```

```python
#Wrapper function
def log_decorator(func):
    def wrapper():
        print(f'calling function: {func.__name__}')
        return func()
    return wrapper

@log_decorator
def say_hello():
    print('Hello!')

say_hello()
```

```python
import time
from pprint import pprint

def log_decorator(func):
    def wrapper(patient_name):
        print(f"Recording patient details: {patient_name}")
        return func(patient_name)
    return wrapper


def time_decorator(func):
    def wrapper(patient_name):
        start = time.time()
        result = func(patient_name)
        end = time.time()
        print(f"Consultation time:\n {end - start:.4f} seconds")
        return result
    return wrapper


@log_decorator
@time_decorator
def doctor_consultation(patient_name):
    pprint(f"Consulting patient: {patient_name}...")
    time.sleep(1)
    print(f"Prescription given to {patient_name}")


doctor_consultation("Alice")
        
        
```

```python
def repeat(times):
    def decorator(func):
        def wrapper(*args, **kwargs):
            for _ in range(times):
                func(*args, **kwargs)
        return wrapper
    return decorator
@repeat(5)
def greet():
    print('hello!')
greet()
```

```python
def check_positive(func):
    def wrapper(num):
        if num > 0:
            print("number is positive")
            print("square is:", func(num))
        else:
            print("number is not positive")
    return wrapper
@check_positive
def square(num):
    return num**2
square(4)
square(-5)
```

```python
#pratice question
def is_even(func):
    def wrapper(num):
        if num%2 == 0:
            print("number is even")
            print("divide_by_2 is:", func(num))
        else:
            print("number is odd")
    return wrapper
@is_even
def divide_by_2(num):
    return num/2
divide_by_2(10)
divide_by_2(7)
```

```python
import time

def time_decorator(func):
    def wrapper():
        start = time.time()
        func()
        end = time.time()
        print(f"Execution time for {func.__name__}: {end - start:.2f} seconds")
    return wrapper

@time_decorator
def long_task():
    for i in range(5):
        print(i)
        time.sleep(1)  # wait 1 second between numbers

long_task()
```
